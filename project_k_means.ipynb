# Customer Segmentation using K-Means

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
# Load data
df = pd.read_csv(r"C:\Users\praveena\Downloads\K_Means\Mall_Customers.csv")
df.head()
```

```python
print(df.info())
print(df.describe())
```

```python
# Visualize Annual Income vs Spending Score
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'])
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Distribution')
plt.show()
```

```python
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X
```

```python
X_scaled
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

print("WCSS values:", wcss)

plt.plot(range(1, 11), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
df.head(10)
```

```python
df
```

```python
plt.figure(figsize=(8,6))
for cluster in range(5):   #0,1,2,3,4
    plt.scatter(
        df[df['Cluster'] == cluster]['Annual Income (k$)'],    # df[df['Cluster'] == 0]['Annual Income (k$)'],
        df[df['Cluster'] == cluster]['Spending Score (1-100)'],  # df[df['Cluster'] == 0]['Spending Score (1-100)'],
        label=f'Cluster {cluster}'
    )

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments')
plt.legend()
plt.show()
```

```python
import joblib

# Save model and scaler
joblib.dump(scaler, "scaler.pkl")
joblib.dump(kmeans, "kmeans_model.pkl")
```
